```python
import pandas as pd

columns = ['title', 'web', 'cite', 'pdf', 'abstract']
CHI_2020 = pd.DataFrame(columns=columns)
```

```python
start_num = 0
article_num = 0

while True:
    url = f'https://scholar.google.com/scholar?start={start_num}&q=source:CHI+source:conference+source:on+source:human+source:factors+source:in+source:computing+source:systems&hl=en&as_sdt=0,5&as_ylo=2020&as_yhi=2020&as_vis=1'

    # 每一面默认十篇文章
    start_num += 10

    # 发送HTTP GET请求获取网页内容
    # response = requests.get(url, proxies=proxy)
    response = requests.get(url, headers=headers, cookies=cookies)

    # 检查响应状态码
    if response.status_code == 200:
        # 使用BeautifulSoup解析网页内容
        soup = BeautifulSoup(response.content, 'html.parser')
        print('请求成功', response.status_code)

    else:
        print('请求失败', response.status_code)

    mid = soup.find(id='gs_res_ccl_mid')
    # 找到根目录下的每个并排元素
    elements = mid.find_all('div', class_='gs_r gs_or gs_scl')

    if len(elements) == 0:
        break

    for element in elements:
        print(f'>>>>>>>>>>>>>>>>>>    Paper {article_num}     <<<<<<<<<<<<<<<<<<<')

        try:
            title_element = element.find('h3', class_='gs_rt')
            link_element = title_element.find('a')

            # 检查是否存在标题和链接元素，若缺失则抛出异常
            if title_element is None or link_element is None:
                raise ValueError("缺少标题或链接元素")

            title = link_element.text
            link = link_element['href']
            print("title:", title)
            print("web:", link)

            pdf_element = element.find('div', class_='gs_or_ggsm')
            if pdf_element is not None:
                pdf_link = pdf_element.find('a')['href']
                if 'pdf' in pdf_link:
                    print("pdf:", pdf_link)
            else:
                pdf_link = 'None'

            cite_element = element.select('div.gs_fl[class="gs_fl"]')
            if len(cite_element) > 0:
                cite_elements = cite_element[0].select('a')

                # 检查是否存在足够的 cite 元素，若缺失则抛出异常
                if len(cite_elements) < 3:
                    raise ValueError("缺少足够的 cite 元素")

                cite = cite_elements[2].text
                if 'by' in cite:
                    str_list = cite.split(' ')
                    cite_num = int(str_list[2])
                    print("cite:", cite_num)
                else:
                    cite_num = 0
            else:
                cite_num = 0

            CHI_2020.loc[article_num, 'title'] = title
            CHI_2020.loc[article_num, 'web'] = link
            CHI_2020.loc[article_num, 'pdf'] = pdf_link
            CHI_2020.loc[article_num, 'cite'] = cite_num

        except ValueError as e:
            # 捕获并处理异常
            print("发生异常:", str(e))

        article_num += 1
```
