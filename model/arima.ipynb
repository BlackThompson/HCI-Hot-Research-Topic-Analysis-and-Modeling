```python
import numpy as np
import matplotlib.pyplot as plt
from pmdarima.arima import auto_arima

# 生成数据
np.random.seed(42)
data = [1, 3, 5, 7, 10, 30, 50, 78, 90]
time_steps = np.arange(1, len(data) + 1)
target_time_steps = np.arange(len(data) + 1, len(data) + 4)  # 预测第10、11、12个数字

# 使用auto_arima函数自动选择ARIMA模型的参数
auto_model = auto_arima(data, start_p=0, start_q=0,
                        max_p=3, max_q=3, m=1,
                        start_P=0, seasonal=False,
                        d=None, D=None, trace=True,
                        error_action='ignore',
                        suppress_warnings=True,
                        stepwise=True)

# 输出自动选择的模型参数
print("自动选择的ARIMA模型参数：", auto_model.order)

# 拟合ARIMA模型
model_fit = auto_model.fit(data)

# 预测第10、11、12个数字
forecast = model_fit.predict(n_periods=3)

# 打印预测结果
print("预测第10、11、12个数字为:", forecast)

# 可视化预测结果
plt.figure(figsize=(8, 5))
plt.plot(time_steps, data, 'o-', label='原始数据')
plt.plot(target_time_steps, forecast, '-o', label='预测数字')
plt.xlabel('时间步长')
plt.ylabel('数据')
plt.legend()

plt.gca().set_facecolor('white')

plt.show()
```
